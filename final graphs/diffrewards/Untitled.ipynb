```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df10_10 = pd.read_csv('re10-10_diff.csv', index_col = 0 )
df10_10_1 = pd.read_csv('re10-10-1_diff.csv', index_col = 0 )
df10_10_1_1 = pd.read_csv('re10-10_1-1_diff.csv', index_col = 0 )
df_diff = pd.read_csv('rediff.csv', index_col = 0 )

y1 = df10_10['me_mean']
y2 = df10_10_1['me_mean']
y3 = df10_10_1_1['me_mean']
y4 = df_diff['me_mean']

error1 = df10_10['me_std']
error2 = df10_10_1['me_std']
error3 = df10_10_1_1['me_std']
error4 = df_diff['me_std']

x = df10_10.index

plt.figure(figsize = (3*2.5,1*4))
plt.plot(x,y1, label = 'RG 1')
plt.fill_between(x,y1-error1, y1+error1, alpha = 0.2 )
plt.plot(x,y3, label = 'RG 2')
plt.fill_between(x,y3-error3, y3+error3, alpha = 0.2)
# plt.plot(x,y2, label = 'G3')
# plt.fill_between(x,y2-error2, y2+error2, alpha = 0.2)
plt.plot(x,y4, label = 'RG 3')
plt.fill_between(x,y4-error4, y4+error4, alpha = 0.2)
plt.title('Effect of reward systems')
plt.xlabel('Round #')
plt.ylabel('Difference in accumulated wins')
plt.legend()
plt.savefig('diff_re1.png')
plt.show()

plt.figure(figsize = (3*5,1*5))
plt.plot(x,y3, label = 'G4')
plt.fill_between(x,y3-error3, y3+error3, alpha = 0.2)
# plt.plot(x,y2, label = 'G3')
# plt.fill_between(x,y2-error2, y2+error2, alpha = 0.2)
plt.plot(x,y1, label = 'G2')
plt.fill_between(x,y1-error1, y1+error1, alpha = 0.2 )
plt.plot(x,y4, label = 'G5')
plt.fill_between(x,y4-error4, y4+error4, alpha = 0.2)
plt.title('Effect of reward systems')
plt.xlabel('Round #')
plt.ylabel('Difference in accumulated wins')
plt.legend()
plt.savefig('diff_re2.png')
plt.show()


# re1 = df10_10.tail(1)
# dif
# re2 = df10_10_1.tail(1)
# re3 = df10_10_1_1.tail(1)
# re4 = df_diff.tail(1)
```

```python
def plot(x,y,error,y1,error1,y2,error2,y3,error3):
    plt.figure(figsize = (15,5))
    plt.fill_between(x,y-error, y+error, alpha = 0.3 , label = '+10,-10')
    plt.plot(x,y)
    plt.fill_between(x,y1-error1, y1+error1, alpha = 0.3  , label = '+10,-10,-1')
    plt.plot(x,y1)
    plt.fill_between(x,y2-error2, y2+error2, alpha = 0.3  , label = '+10,-10,+1,-1')
    plt.plot(x,y2)
    plt.fill_between(x,y3-error3, y3+error3, alpha = 0.3  , label = 'diff')
    plt.plot(x,y3)
    plt.title('Effect of learning rate')
    plt.xlabel('Game #')
    plt.ylabel('Accumulated wins')
    plt.legend()
    plt.savefig('diff_lr.png')
    plt.show()
```

```python
def plot2(x,y,error,y1,error1 , title , img):
    plt.figure(figsize = (15,5))
    plt.fill_between(x,y-error, y+error, alpha = 0.3 , label = 'Opponent Modeling')
    plt.plot(x,y)
    plt.fill_between(x,y1-error1, y1+error1, alpha = 0.3  , label = 'Minimax')
    plt.plot(x,y1)
    plt.title(title)
    plt.xlabel('Game #')
    plt.ylabel('Accumulated wins')
    plt.legend()
    plt.savefig(img)
    plt.show()
```

```python
df10_10 = pd.read_csv('re10-10.csv', index_col = 0 )
df10_10_1 = pd.read_csv('re10-10-1.csv', index_col = 0 )
df10_10_1_1 = pd.read_csv('re10-10_1-1.csv', index_col = 0 )
df_diff = pd.read_csv('rediff.csv', index_col = 0 )

plot2(df10_10.index,df10_10['me_mean'],df10_10['me_std'],df10_10['opp_mean'],df10_10['opp_std'], '10,-10 reward' , '1010.png')
plot2(df10_10_1.index,df10_10_1['me_mean'],df10_10_1['me_std'],df10_10_1['opp_mean'],df10_10_1['opp_std'], '10,-10,-1 reward' , '10101.png')
plot2(df10_10_1_1.index,df10_10_1_1['me_mean'],df10_10_1_1['me_std'],df10_10_1_1['opp_mean'],df10_10_1_1['opp_std'], '10,-10,1,-1 reward' , '101011.png')
plot2(df_diff.index,df_diff['me_mean'],df_diff['me_std'],df_diff['opp_mean'],df_diff['opp_std'], 'different reward' , 'diff.png')

# plot(df10_10.index,df10_10['me_mean'],df10_10['me_std'],df10_10_1['me_mean'],df10_10_1['me_std'],df10_10_1_1['me_mean'],df10_10_1_1['me_std'],df_diff['me_mean'],df_diff['me_std'])
```
