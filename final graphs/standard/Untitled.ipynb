```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mm = pd.read_csv('mm_diff.csv', index_col = 0 )
# om = pd.read_csv('difflr08.csv', index_col = 0 )
both = pd.read_csv('difflr08.csv', index_col = 0 )

y1 = mm['me_mean']
# y2 = om['me_mean']
y3 = both['me_mean']

error1 = mm['me_std']
# error2 = om['me_std']
error3 = both['me_std']

x = both.index

plt.figure(figsize = (3*2.5,1*4))
plt.plot(x,y1, label = 'G 1')
plt.fill_between(x,y1-error1, y1+error1, alpha = 0.2 )
plt.plot(x,y3, label = 'G 2')
plt.fill_between(x,y3-error3, y3+error3, alpha = 0.2)
# plt.plot(x,y2, label = 'om')
# plt.fill_between(x,y2-error2, y2+error2, alpha = 0.2)

plt.title('Opponent Modelling Q-learning final results')
plt.xlabel('Round #')
plt.ylabel('Difference in accumulated wins')
plt.legend()
plt.savefig('stand.png')
plt.show()
```
