```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot(x,y,error,y1,error1,y2,error2):
    plt.figure(figsize = (15,5))
    plt.fill_between(x,y-error, y+error, alpha = 0.3 , label = 'alpha = 0.3')
    plt.plot(x,y)
    plt.fill_between(x,y1-error1, y1+error1, alpha = 0.3  , label = 'alpha = 0.5')
    plt.plot(x,y1)
    plt.fill_between(x,y2-error2, y2+error2, alpha = 0.3  , label = 'alpha = 0.8')
    plt.plot(x,y2)
    plt.title('Effect of learning rate')
    plt.xlabel('Game #')
    plt.ylabel('Accumulated wins')
    plt.legend()
    plt.savefig('diff_lr.png')
    plt.show()
```

```python
df_03 = pd.read_csv('difflr03.csv', index_col = 0 )
df_05 = pd.read_csv('rediff.csv', index_col = 0 )
df_08 = pd.read_csv('difflr08.csv', index_col = 0 )

plot(df_03.index,df_03['me_mean'],df_03['me_std'],df_05['me_mean'],df_05['me_std'],df_08['me_mean'],df_08['me_std'])
```
