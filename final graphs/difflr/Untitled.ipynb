```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
lr5 = pd.read_csv('rediff.csv', index_col = 0 )
lr3 = pd.read_csv('difflr03.csv', index_col = 0 )
lr8 = pd.read_csv('difflr08.csv', index_col = 0 )

y1 = lr5['me_mean']
y2 = lr3['me_mean']
y3 = lr8['me_mean']

error1 = lr5['me_std']
error2 = lr3['me_std']
error3 = lr8['me_std']

x = lr5.index

plt.figure(figsize = (3*2.5,1*4))
plt.plot(x,y2, label = 'LG 1')
plt.fill_between(x,y2-error2, y2+error2, alpha = 0.2)
plt.plot(x,y1, label = 'LG 2')
plt.fill_between(x,y1-error1, y1+error1, alpha = 0.2 )
plt.plot(x,y3, label = 'LG 3')
plt.fill_between(x,y3-error3, y3+error3, alpha = 0.2)

plt.title('Effect of learning rate in OM')
plt.xlabel('Round #')
plt.ylabel('Difference in accumulated wins')
plt.legend()
plt.savefig('diff_lr.png')
plt.show()
```
